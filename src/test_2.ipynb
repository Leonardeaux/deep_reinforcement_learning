```python
import gym
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

env = gym.make('CartPole-v1')
state_size = env.observation_space.shape[0]
action_size = env.action_space.n

state_size
```

```python
def build_model(state_size, action_size):
    model = Sequential()
    model.add(Dense(24, input_dim=state_size, activation='relu'))
    model.add(Dense(24, activation='relu'))
    model.add(Dense(action_size, activation='linear'))
    model.compile(loss='mse', optimizer=Adam(lr=0.001))
    return model

model = build_model(state_size, action_size)
```

```python
def choose_action(state, epsilon):
    if np.random.rand() <= epsilon:
        return env.action_space.sample()
    q_values = model.predict(state)
    return np.argmax(q_values[0])
```

```python
def train_model(model, state, action, reward, next_state, done):
    target = reward
    if not done:
        target = reward + 0.95 * np.amax(model.predict(next_state)[0])
    target_f = model.predict(state)
    target_f[0][action] = target
    model.fit(state, target_f, epochs=1, verbose=0)
```

```python
episodes = 30

scores = []

for e in range(episodes):
    state = env.reset()[0]
    state = np.reshape(state, [1, state_size])
    done = False
    score = 0
    
    while not done:
        # env.render()  # Uncomment this to see the simulation
        action = choose_action(state, epsilon=0.2)
        next_state, reward, done, _, _ = env.step(action)
        next_state = np.reshape(next_state, [1, state_size])
        
        score += reward
        train_model(model, state, action, reward, next_state, done)
        
        state = next_state
    
    print(f"Episode {e+1}/{episodes}, score: {score}")
    scores.append(score)
    
average_score = np.mean(scores)
print(f"Moyenne des scores sur {episodes} pisodes: {average_score}")
```
